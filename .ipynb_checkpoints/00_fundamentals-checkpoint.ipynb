# Tensor

## Creating Tensors
Tensors are created using `torch.tensor()`

By convention `Scaler` & `Vector` are created with lower-case variable names e.g. `x, y, scl, vct`

`Matrix` & `Tensor` are created with upper-case variable names e.g. `X, Y, MTX, TNSR`

```python
scaler = torch.tensor(7)

print(f"{scaler=}") # Tensor Rank-0
print(f"{scaler.ndim=}") # Rank/N-Dimension
print(f"{scaler.shape=}") # Tensor Shape [Rows * Columns]
print(f"{scaler.item()=}") # Tensor in Python int
```

```python
vector = torch.tensor([7, 7])

print(f"{vector=}") # Tensor Rank-1
print(f"{vector.ndim=}")
print(f"{vector.shape=}")
```

```python
MATRIX = torch.tensor([[7, 8, 9],
                       [9, 10, 11]])
print(f"{MATRIX=}")
print(f"{MATRIX[0]=}")
print(f"{MATRIX[1]=}")
print(f"{MATRIX[0][1]=}")
print(f"{MATRIX.ndim=}")
print(f"{MATRIX.shape=}")
```

## Random Tensors

Random Tensors are tensors of a particular shape filled with random values. 

**Need for a Random Tensor:**

Many NNs learn in a way that they start with tensors full of random numbers, and they adjust those tensors as the learning progresses.

`Repeat(Start with Random Tensor -> Look at Data -> Update Random Tensor)`

```python
RND_T = torch.rand(size=(3, 4)) ## Size/Shape

print(f"{RND_T=}")
print(f"{RND_T.ndim=}")
print(f"{RND_T.shape=}")
```

### Multiplication

There are 2 ways to perform multiplication

1. Element-Wise
2. Matrix Multiplication (Dot Product)

There are two rules for matrix Multiplication:
1. The **inner dimensions** must match: number of columns of T1 should be equal to number of rows of T2
   - `(3, 2) @ (3, 2)` won't Work
   - `(3, 2) @ (2, 3)` will Work
   - `(2, 3) @ (3, 2)` will Work
2. The resulting matrix has the shape of the **outer dimensions**: Number of rows in T1 and number of Columns in T2
   - `(3, 2) @ (2, 3)` -> `(3, 3)`
   - `(2, 3) @ (3, 2)` -> `(2, 2)`

```python
# Element Wise
print(f"{tsr=}")
print(f"{tsr} * {tsr}", end=" = ")
print(f"{tsr*tsr}")
```

```python
tsr @ tsr
```

```python
tsr = torch.arange(10000000)
tsr
```

```python
%%time
value = 0
for i in range(len(tsr)):
    value += tsr[i] * tsr[i]
print(value)
```

```python
%%time
tsr @ tsr
```

## Reshaping, Stacking, Squeezing and Unsqueezing Tensors

- Reshaping: Reshape an input tensor to a defined/required shape
- View: Return a view of an input tensor of certain shape but keep the same memory as the original tensor
- Stacking: combine multiple tensors on top of each other (V-Stack) or side-by-side (H-Stack)
- Squeeze: Removes all `1` dimensions from a tensor
- Unsqueeze: add a `1` dimensions to a tensor
- Permute: Return a view of the input width dimensions permuted (swapped) in a certain way

```python
x = torch.arange(1., 10.)
x, x.shape
```

```python
, # Reshape: Add an extra dimension
# The new shape needs to be compatible with the existing shape.
# Basically the nyumber of tensors in the source tensors should fit exactly in the new shape
# the New shapes multiplication should be equal to original shape
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
z = x.view(1, 9)
z, z.shape
```

```python
x[5] = 100
x, x_reshaped, z
```

```python
z[:, 0] = 20
x, x_reshaped, z
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

```python
# Stack tensors side-by-side of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
torch.vstack([x, x, x, x])
```

```python
torch.hstack([x.view(9, 1), x.view(9, 1), x.view(9, 1), x.view(9, 1)])
```

## Pytorch Tensors & NumPy

NumPy is popular scientific Python numerical computing library.

Pythorch has functionality to interact with it.

- Data in NumPy to PyTorch Tensor: `torch.from_numpy(ndarray)`
- Pytorch Tensor to NumPy: `torch.tensor.numpy()`

```python
# NumPy to Tensor
arr = np.arange(1., 8.)
tsr = torch.from_numpy(arr) # Default dtype of numpy is float64 and tensor default dtype if float32. So tensor reflects that

arr, tsr
```

```python
# Change value of array
arr = arr + 1
arr, tsr
```

```python
# Tensor to NumPy
tsr = torch.ones(10)
arr = tsr.numpy()
tsr, arr
```

```python
tsr += 1
tsr, arr
```

```python
tsr[0] = 10
tsr, arr
```

```python
tsr += 1
tsr, arr
```

```python
tsr = tsr + 1
tsr, arr
```

## Reproducibility

For reproducibility, we need consistency across devices & different runs of the code. This can be done by turning off the randomness.

```python
import torch

A = torch.rand(3, 4)
B = torch.rand(3, 4)

A, B, (A == B).all()
```

```python
# Set random seed
torch.manual_seed(42)
```

```python
torch.manual_seed(42)
C = torch.rand(3, 4)

torch.manual_seed(42)
D = torch.rand(3, 4)

C, D, (C == D).all()
```

## Putting tensors (and models) on the GPU

The reason we want our tensors/models on the GPU is because using a GPU results in faster computations.

```python
# Create a tensor (defaults to CPU)
tensor1 = torch.tensor([1, 2, 3])
tensor1, tensor1.device
```

```python
# Create a tensor (defaults to CPU)
tensor2 = torch.tensor([1, 2, 3], device=device)
tensor2, tensor2.device
```

```python
# Move Tensor to a particular device. This creates a copy of the source tensor in the selected device.
tensor3 = tensor1.to(device)
tensor3, tensor3.device
```

```python
tensor1[0] = 10
tensor1, tensor3
```
